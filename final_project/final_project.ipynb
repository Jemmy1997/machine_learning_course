```python
dataFrame = pd.read_csv('fertility.csv')
dataFrame["Frequency of alcohol consumption"].unique()
```

```python
dataFrame['Season'].replace({"winter":2, "spring":1, "summer":-2, "fall":-1 }, inplace= True)
dataFrame['Childish diseases'].replace({"yes":1, "no":0}, inplace= True)
dataFrame['Accident or serious trauma'].replace({"yes":1, "no":0}, inplace= True)
dataFrame['Smoking habit'].replace({"daily":2, "occasional":1,"never":0}, inplace= True)
dataFrame['Frequency of alcohol consumption'].replace({"once a week":4,"hardly ever or never":3,"several times a week":2,"several times a day":1, "every day":0 }, inplace= True)
dataFrame['Surgical intervention'].replace({"yes":1,"no":0}, inplace= True)
dataFrame['Diagnosis'].replace({"Normal":1,"Altered":0}, inplace= True)
dataFrame['High fevers in the last year'].replace({"more than 3 months ago":2,"less than 3 months ago":1,"no":0}, inplace= True)
dataFrame
```
